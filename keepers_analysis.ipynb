```python
import pandas as pd
```

```python
keepers = pd.read_csv('goalkeepers.csv')
```

```python
keepers
```

```python
# Analyzing players by height
```

```python
height_average = round(keepers['Height'].mean(),2)
```

```python
height_average
```

```python
keepers[keepers['Nationality'] == 'Brazil' ]['Height'].mean()
```

```python
nat = list(keepers['Nationality'].unique())
```

```python
nat
```

```python
# average height by nationality 

average = {}
for n in nat:
    average[n] = round(keepers[keepers['Nationality'] == n ]['Height'].mean(), 2)
```

```python
average = pd.Series(average, name='Average Height')
```

```python
nationalities = keepers['Nationality'].value_counts()
```

```python
average
```

```python
nationalities
```

```python
country = pd.DataFrame([average, nationalities.rename('Players Count')])
```

```python
country.plot.bar()
```

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.animation as animation

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))


def animate(i):
    line.set_ydata(np.sin(x + i / 50))  # update the data.
    return line,


ani = animation.FuncAnimation(
    fig, animate, interval=20, blit=True, save_count=50)

# To save the animation, use e.g.
#
# ani.save("movie.mp4")
#
# or
#
# writer = animation.FFMpegWriter(
#     fps=15, metadata=dict(artist='Me'), bitrate=1800)
# ani.save("movie.mp4", writer=writer)

plt.show()
```
